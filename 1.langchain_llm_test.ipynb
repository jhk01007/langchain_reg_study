3. LangChain 써보기

```python
from langchain_openai import ChatOpenAI
import os

api_key = os.getenv("OPENAI_API_KEY") # api key를 OPENAI_API_KEY로 환경변수에 저장하면 이 부분을 생략할 수 있다.

llm = ChatOpenAI(api_key=api_key) # 변수명은 llm으로 쓰는 것이 관례
```

```python
ai_message = llm.invoke("인프런에 어떤 강의가 있나요?")
ai_message
```
